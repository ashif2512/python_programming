# **7. Nested Lists**

```
n = int(input()) 

name_list = []
mark_list = []

for i in range(n):
    name_list.append(input())
    mark_list.append(input())

min_mark = sorted(mark_list)[1:2] 
outer_list = []

for name in range(len(name_list)):
	inner_list = []
	inner_list.append(name_list[name])
	inner_list.append(mark_list[name])
	outer_list.append(inner_list)

lowest_number_name_list = []
for i in range(len(outer_list)):
    if mark_list[i] == min_mark[0]:
        
        lowest_number_name_list.append(name_list[i])

sorted_list = sorted(lowest_number_name_list, key=str.lower)
for i in range(len(sorted_list)):
    print(sorted_list[i])
    
```

# **5. Basic Statistics Warmup**

```
n = int(input("Input your number of quantity from user:")) 
numbers = input("Enter your values from user:  ").split()
numbers = sorted([int(x) for x in numbers])

def mean(numbers):
    size = len(numbers)
    return round(sum(numbers)/float(size),1)

def median(numbers,size):
    size = len(numbers)
    mid1 = int(round(size/2)) - 1 + n%2 
    mid2 = int(round(size/2))
    return round(mean([numbers[mid1],numbers[mid2]]),1)

def mode(numbers):
    return max(numbers, key=numbers.count)

def st_dev(numbers):
    total = 0
    mean_nums = mean(numbers)
    size = len(numbers)
    for number in numbers:
        total += (number - mean_nums)**2
    return round((total / size)**0.5,1)

def confidence(numbers): 
    s = 1.96 
    size = len(numbers)
    stderr = st_dev(numbers)/(size**0.5) 
    mean_nums = mean(numbers)
    lower = round(mean_nums - (stderr * s),1)
    upper = round(mean_nums + (stderr * s),1)
    return [lower, upper]


print ("Mean = ", mean(numbers))
print ("Median = ", median(numbers,n))
print ("Mode = ", mode(numbers))
print ("Standard deviation = ", st_dev(numbers))
for item in confidence(numbers): 
    print (item)
```
